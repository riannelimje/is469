```python
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
# Load your fine-tuned model and tokenizer
model_name = "distilgpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
# Define a prompt
prompt = "The quick brown fox jumps over the"
```

```python
# Encode the prompt
input_ids = tokenizer.encode(prompt, return_tensors="pt")
```

```python
#  Let's experiment with the impact of changing temperature, in the output softmax, on generated text
# A lower temperature (e.g., 0.7) makes the output more deterministic
# A higher temperature (e.g., 1.0 or higher) makes the output more random
output = model.generate(
    input_ids,
    max_length=50,
    num_return_sequences=1,
    temperature=0.0001,  # Set the temperature here
    do_sample=True,   # Ensure sampling is enabled for temperature to have an effect
)
```

```python
# Let's generate !
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print(generated_text)
```
