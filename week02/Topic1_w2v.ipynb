```python
from gensim.models.word2vec import Word2Vec
import gensim.downloader as api
from gensim.utils import simple_preprocess
import gensim.models
```

```python
wv = api.load('20-newsgroups') # for a smaller dataset faster to download you may use instead : 'semeval-2016-2017-task3-subtaskBC'
# download the corpus and return it opened as an iterable
# other datasets available at - https://github.com/piskvorky/gensim-data

# Extract text data from the 'data' field and preprocess
processed_text = [simple_preprocess(doc['data']) for doc in wv]

model = Word2Vec(processed_text, min_count=3)  # train a model from the corpus
```

```python
vocab = list(model.wv.key_to_index.keys())
print(vocab)
print("Vocab len is",len(vocab))
```

```python
model.wv.most_similar("question")
```

```python
model.wv.similarity('question', 'university')
```

```python
model.wv.most_similar(positive=['sentence'], topn=5)
```

```python
print(model.wv.doesnt_match(['dog', 'cat', 'pig', 'horse', 'university', 'goat']))
```
